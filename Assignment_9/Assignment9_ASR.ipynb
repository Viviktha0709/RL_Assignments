```python
import speech_recognition as sr

rec = sr.Recognizer()

print("Speak something...")

with sr.Microphone() as source:
    audio = rec.listen(source)

try:
    text = rec.recognize_google(audio)
    print("Recognized Speech:", text)
except Exception as e:
    print("Error:", e)
```
